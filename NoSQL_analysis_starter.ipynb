### 1. Which establishments have a hygiene score equal to 20?

```python
# Find the establishments with a hygiene score of 20
query = {'scores.Hygiene': {'$eq': 20}}

# Use count_documents to display the number of documents in the result
results_hygene = establishments.find(query)

# Display the first document in the results using pprint
print("total establishments with Hygene score 20 is:", establishments.count_documents(query))
```

```python
# Convert the result to a Pandas DataFrame
Hygene_df = DataFrame(results_hygene)
# Display the number of rows in the DataFrame
Hygene_df.info()
```

### 2. Which establishments in London have a `RatingValue` greater than or equal to 4?

```python
# Find the establishments with London as the Local Authority and has a RatingValue greater than or equal to 4.

query = {"LocalAuthorityName": {'$regex': "London"}, "RatingValue": {"$gte": 4}} 

# Use count_documents to display the number of documents in the result
results_Localauthority = establishments.find(query)

# Display the first document in the results using pprint
for i in range(1):
 pprint(results_Localauthority[i])


# Use count_documents to display the number of documents in the result
test = establishments.count_documents(query)

# Display the first document in the results using pprint
print("Total of establishments with London as the Local Authority:", test)
```

### 3. What are the top 5 establishments with a `RatingValue` rating value of 5, sorted by lowest hygiene score, nearest to the new restaurant added, "Penang Flavours"?

```python
# Search within 0.01 degree on either side of the latitude and longitude.

# Construct the query
query = {
    "$and": [
        {"geocode.longitude": {"$gte": 0.08384000, "$lte": 0.09384}},
        {"geocode.latitude": {"$gte": 51.49014200, "$lte": 51.500142}},
        {"RatingValue": 5}
    ]
}

# Define the sorting criteria
sort = [('scores.Hygiene', 1)]

# Define the limit for the number of results
limit = 5

# Perform the query, sort, and limit the results
results_top5 = establishments.find(query).sort(sort).limit(limit)

# Convert the cursor to a list of dictionaries
results_list = list(results_top5)

# Convert the list of dictionaries to a DataFrame
top_5_near_df = pd.DataFrame(results_list)

# Display the information about the DataFrame
top_5_near_df.info()

# Display the DataFrame
print(top_5_near_df)
```

```python
# Convert result to Pandas DataFrame
top_5_near_df.head()
```

### 4. How many establishments in each Local Authority area have a hygiene score of 0?

```python
# Create a pipeline that: 
# 1. Matches establishments with a hygiene score of 0
pipeline = [
    {
        "$match": {
            "scores.Hygiene": 0 
        }
    },
    {
        "$group": {
            "_id": "$LocalAuthorityName",
            "total_establishments": {"$sum": 1}
        }
    },
    {
        "$sort": {"total_establishments": -1}
    },
    {
        "$limit": 10
    }
]

# Perform the aggregation query
results_scorezero = establishments.aggregate(pipeline)

# Print the results
pprint(list(results_scorezero))

aggregation_results = [
    {'_id': 'Thanet', 'total_establishments': 1130},
    {'_id': 'Greenwich', 'total_establishments': 882},
    {'_id': 'Maidstone', 'total_establishments': 713},
    {'_id': 'Newham', 'total_establishments': 711},
    {'_id': 'Swale', 'total_establishments': 686},
    {'_id': 'Chelmsford', 'total_establishments': 680},
    {'_id': 'Medway', 'total_establishments': 672},
    {'_id': 'Bexley', 'total_establishments': 607},
    {'_id': 'Southend-On-Sea', 'total_establishments': 586},
    {'_id': 'Tendring', 'total_establishments': 542}
]

# Convert the list of dictionaries to a DataFrame
scorezero_df = pd.DataFrame(aggregation_results)

# Display the DataFrame
print(scorezero_df)
```
